```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

# Criação das listas para o pandas

dados_precos = []
dados_links = []

# Criação do navegador automático do Selenium e Entrando no site da amazon

navegador = webdriver.Chrome()
navegador.get("https://www.amazon.com.br")
time.sleep(3)

# Digitando a categoria na barra de pesquisas e apertando em pesquisar

navegador.find_element(By.XPATH, '//*[@id="twotabsearchtextbox"]').send_keys('Roupas Femininas Para Frio')
navegador.find_element(By.XPATH, '//*[@id="nav-search-submit-button"]').click()
time.sleep(3)

# Criando as variáveis das listas para poderem ser percorridas

lista_roupas = navegador.find_elements(By.CSS_SELECTOR, '.a-link-normal.s-no-outline')
lista_precos = navegador.find_elements(By.CSS_SELECTOR, '.a-price .a-offscreen')

# Percorrendo as litas e adicionando seus outputs nas listas em pandas

for roupas in lista_roupas:
    link = roupas.get_attribute('href')
    dados_links.append(link)

for precos in lista_precos:
    valor = precos.get_attribute('textContent')
    dados_precos.append(valor)

# Opção do pandas para visualização completa dos links na tabela

pd.set_option('display.max_colwidth', None)

# Criação e Display na tabela do pandas

print('\n ------- TABELA DOS PRODUTOS ------- ')
df = pd.DataFrame(list(zip(dados_links, dados_precos)), columns =['Link do Produto', 'Preço'])
display(df)
```
